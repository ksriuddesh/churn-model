```python
#importing libraries 

import pandas as pd 
import numpy as np 
import tensorflow as tf
from  tensorflow.keras.models import load_model
import pickle 
from sklearn.preprocessing import OneHotEncoder

```

```python
#load the model 
model = load_model('model.h5')
with open('gender_lable_encoder.pkl','rb') as file:
    lable_encoder_gender = pickle.load(file)

with open('OneHotEncoder_geaography.pkl','rb') as file:
    geo_encoder = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler_encoder = pickle.load(file)
```

```python
input_data = {
    
    
    
    'CreditScore': 619,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 42,
    'Tenure': 3,
    'Balance': 70000,
    'NumOfProducts': 1,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 101348.88,
    
}
```

```python
geo_encoded = geo_encoder.transform([[input_data['Geography']]])
geo_encoded_array = geo_encoded.toarray()
geo_encoded_df = pd.DataFrame(geo_encoded_array, columns=geo_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] = lable_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
input_df = pd.concat([input_df.drop("Geography", axis=1), geo_encoded_df], axis=1)
```

```python
input_df
```

```python
input_scaled = scaler_encoder.transform(input_df)
input_scaled
```

```python
predict_model = model.predict(input_scaled)
predict_model
```

```python
prediction_prob = predict_model[0][0]

prediction_prob
```

```python
if prediction_prob > 0.5:
    print("the customer is likey to stop usoing the account")
elif prediction_prob<0.5:
    print("the customer will be in the game ")
```
